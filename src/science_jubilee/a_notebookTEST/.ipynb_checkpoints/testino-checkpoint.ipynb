```python
# On récupère le dossier actuel (où se trouve le Notebook)
current_dir = os.getcwd()

# On cherche la racine du projet "science-jubilee" dans le chemin
if "science-jubilee" in current_dir:
    # On coupe le chemin juste après "science-jubilee"
    project_root = current_dir.split("science-jubilee")[0] + "science-jubilee"
else:
    # Si on ne trouve pas le nom du dossier, on reste sur le dossier actuel
    project_root = current_dir

#nom du fichier cible
nom_fichier = "plan_jubilee.txt"
plan = os.path.join(project_root, "src", "science_jubilee", "decks", "deck_definition", nom_fichier)

print(f"Début du tracé : {plan}")

if not os.path.exists(plan):
    print(f"ERREUR : Le fichier est introuvable à l'adresse : {plan}")
else:
    with open(plan, 'r', encoding='utf-8') as f:
        #nettoyage ligne
        lignes = [l.strip() for l in f.readlines() if l.strip() and not l.startswith(';')]
        total = len(lignes)

    print(f"Envoi de {total} commandes au contrôleur...")
    
    for i, commande in enumerate(lignes):
        jm.controller.gcode(commande)
        
        # Affichage de la progression toutes les 10 lignes
        if i % 10 == 0:
            print(f"Progression : {i}/{total} commandes envoyées...")

    print("✅ Terminé ! Le tracé est fini.")
```

```python
# 1. On récupère les infos de ton outil par son nom
dict_outil = jm.get_tool_by_name("Inoculator")
idx = dict_outil['index']

# 2. On prépare l'outil physiquement
jm.controller.pickup_tool_sequence(idx)
jm.set_active_tool(idx)
jm.set_tool_offset(idx, (0, -31.5, 35))

tool0 = jm.tools_list[idx]
tool0.__class__ = Loop
tool0._machine = jm.controller  # On le lie au contrôleur pour les mouvements
tool0.is_active_tool = True     # On valide le décorateur de sécurité

# 4. CORRECTIF : On crée la méthode safe_z_movement si elle manque
if not hasattr(jm.controller, 'safe_z_movement'):
    # On remonte à 80mm de hauteur pour ne rien toucher
    tool0.safe_z_movement = lambda: jm.controller.move_to(z=80)
    print("✅ Correctif safe_z_movement appliqué.")

print(f"✅ Outil '{tool0.name}' prêt (Classe: {type(tool0).__name__})")
```

```python
rows = ['A', 'B', 'C', 'D']
cols = range(1, 7)

# On définit la SOURCE fixe (ici slot A1 du slot 
slot_0 = jm.deck.get_slot('0')

"""
well_source = slot_0.labware.wells['A1']
well_source.y = well_source.y - 36 + slot_0.coordinates[1] #Offset inoculateur
well_source.x = well_source.x + slot_0.coordinates[0]#maj x
print(well_source)
"""


print(f"🚀 Début de la distribution depuis {well_source.name}...")

# --- LA BOUCLE DE DISTRIBUTION ---
for r in rows:
    for c in cols:
        well_dest_name = f"{r}{c}" # Génère A1, A2... D6

        
        # On récupère l'objet Well de destination
        well_dest = slot_0.labware.wells[well_dest_name]

        well_dest.x = (well_dest.x + slot_0.coordinates[0])
        well_dest.y = (well_dest.y-36 + slot_0.coordinates[1])#-36 ici car transferyt utilise move_to et pas effecteur to


        print(f"Transfert de {well_source.name} vers {well_dest_name}...")
        jm.controller.move_to(x=well_dest.x, y=well_dest.y+36)
        jm.controller.move_to(z=50)
        detect.capture_octopi_image("avant")
        jm.controller.move_to(z=150)

        while(detect.detect_diff() == False):
            try:
                # Appel du transfert (la sécurité safe_z est déjà injectée)
                tool0.transfer(
                    source=well_source,      # FIXE
                    destination=well_dest,   # CHANGE à chaque itération
                    s=2000,
                    sweep_x=3,
                    sweep_y=3,
                    sweep_z=10,
                    sweep_speed=150,
                    up_speed=800,
                    randomize_pickup=True
                )
            except Exception as e:
                print(f"❌ Erreur critique au puits {well_dest_name}: {e}")
                # Sécurité : on remonte immédiatement
                jm.controller.move_to(z=80)
                break 
            jm.controller.move_to(z=50)
            jm.controller.move_to(x=well_dest.x, y=well_dest.y+36)
            detect.capture_octopi_image("après");


print("✅ Distribution terminée sur toute la plaque !")
```
